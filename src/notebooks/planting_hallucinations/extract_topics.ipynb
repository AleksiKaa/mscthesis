```python
import os
import sys

sys.path.append(
    os.path.abspath(os.path.join(os.getcwd(), "../.."))
)  # Add parent directory to path

import json
import pandas as pd
```

```python
# Load original dataset
orig = r"../../../data/complete_dataset.csv"
original_dataset = pd.read_csv(orig, sep=";")

theme_to_topics = {}
for i, row in original_dataset.iterrows():
    theme = row["theme"]
    topic = row["topic"]

    if theme_to_topics.get(theme, None) is None:
        theme_to_topics[theme] = set()

    theme_to_topics[theme].add(topic)

theme_to_topics
```
